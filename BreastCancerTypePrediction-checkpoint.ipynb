```python
cr.diagnosis.replace({"B":0, "M":1}, inplace = True)
```

```python
cr.diagnosis = cr.diagnosis.astype('category')
```

```python
cr_x = cr.iloc[:, 2:-1] #All Independent Features
```

```python
cr_x.head()
```

```python
cr_y = cr.iloc[:, 1] #Dependent Features
```

```python
cr_y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
cr_x_train, cr_x_test, cr_y_train, cr_y_test = train_test_split(cr_x, cr_y, test_size = .2, random_state = 101)
```

```python
cr_x_train.shape
```

```python
cr_y_train.shape
```

```python
cr_x_test.shape
```

```python
cr_y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#Instance of LogisticRegression model
logmodel = LogisticRegression() 
```

```python
logmodel.fit(cr_x_train, cr_y_train)
```

```python
pred = logmodel.predict(cr_x_test)
```

```python
pred
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
```

```python
confusionMatrix = confusion_matrix(pred,cr_y_test)
confusionMatrix
```

```python
tP = confusionMatrix[0][0] #truePositive
tN = confusionMatrix[1][1] #trueNegative
fP = confusionMatrix[1][0] #falsePositive
fN = confusionMatrix[0][1] #falseNegative
print(tP)
```

```python
accuracy = (tP + tN) / (tP + tN + fP + fN)
accuracy
```

```python
#Prediction with probability
pred_prob = logmodel.predict_proba(cr_x_test)
pred_prob

#In each row 1st column tells the probability of output value
#being 0 and 2nd column tells the probability of output value
#being 1
```
